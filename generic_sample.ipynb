<h1>Generic Sample of Linnear Regresion</h1>

<h2>Import the librarys to use.</h2>

```python
import numpy as np #Numeric Library
import matplotlib.pyplot as plt #Generate grafics 
```

<h2>Import the datasets</h2>

```python
from sklearn.linear_model import LinearRegression
data = LinearRegression()

from sklearn.metrics import mean_squared_error
```

<h2>Here we start to code the function, calculate the value of Y and plot the points generated by the values of X.</h2>

```python
def func(x): # function f(x) = 0.1*x + 1.25 + 0.2*Gaussian_noise
    np.random.seed(42)
    y = 0.1*x + 1.25 + np.random.randn(x.shape[0])#The last part is for generate Gaussian Nosie in the funcion

    return y

x = np.arange(0, 200, 0.5) #It generates values of X from 0 to 20 in intervals of .5
y = func(x) #And then we caculate the function we've generated

#We call plt to generete an image of the points that we're gonna use
plt.scatter(x,y,label='data', color='blue', alpha=0.2)
plt.title('Data');
```

<h2>Here we teach the model whit the data in X and Y</h2>

```python
# we instruct linear regression to learn from the data (x, y)
data.fit(x.reshape(-1,1),y)

print('w = '+str(data.coef_) + ' and b = ' + str(data.intercept_))
```

<h2>Theses lines of code calculate the MSE and RMSE</h2>

```python
new_x = np.array([5])
#This is the prediction for Y whit X = 5 in the model we've trained
prediction = data.predict(new_x.reshape(-1,1))

#This is the prediction for all the values of Y according whit all the values of X
predict_training = data.predict(x.reshape(-1,1))

#Here the program caculates the MSE based on the real values of Y and the predicted values of Y
mse = mean_squared_error(y_true=y,y_pred=predict_training)

#Here we calculate the root of MSE
rmse = np.sqrt(mse)

print('The Mean Squared Error is:' + str(mse))

print('The root of the Mean Squared Error is:' + str(rmse))
```

<h2>In this part of the code we calculate the Determination coefficient R2</h2> 

```python
#Whit the method score we calculate R2 where 1 is ok and 0 is wrong
r2 = data.score(x.reshape(-1,1),y)
print('The etermination coefficient R2 is: ' + str(r2))
```

<h2>Finally we plot the results of our calculus</h2>

```python
plt.scatter(x,y,label='data', color='blue', alpha=0.2)

w = data.coef_
b = data.intercept_

plt.plot([0,200],[b, b + w*200], c="red")

plt.title('Linnear regresion');
plt.show()
```
